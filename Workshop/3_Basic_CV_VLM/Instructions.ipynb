        - In JupyterLabs File Browser, choose the folder where you want to clone the GitHub repository (e.g., the home folder).
        - Select Git > Clone a Repository and paste the HTTPS URL for your GitHub repository.
        - If prompted, enter your credentials (GitHub username and personal access token/password if using two-factor authentication) - You will need to generate one via Github > settings > developer > classic token.
        - Click Clone, and your Vertex AI Workbench instance will show your repository as a new folder.
